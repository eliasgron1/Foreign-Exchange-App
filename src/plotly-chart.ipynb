```python
import requests
import defs, utils
import oanda_api
import pandas as pd
import plotly.graph_objects as go

# This file Will draw a price candle chart of the given pair and granularity


pair = "EUR_USD"
gran = "H1"

df = pd.read_pickle(utils.get_historic_candles_filename(pair))

ignore_cols = ['time', 'volume']

mod_cols = [x for x in df.columns if x not in ignore_cols]

df[mod_cols] = df[mod_cols].apply(pd.to_numeric)

df_plot = df.iloc[-100:].copy()


figure = go.Figure()
figure.add_trace(go.Candlestick(
    x=df_plot.time, open=df_plot.mid_o, high=df_plot.mid_h, low=df_plot.mid_l, close=df_plot.mid_c,
    line=dict(width=1), opacity=1
))

figure.update_layout(
    width=1400,
    height=600,
    margin=dict(l=10,r=10,b=10,t=10),
    font=dict(size=10, color="#e1e1e1"),
    paper_bgcolor='#1e1e1e',
    plot_bgcolor="#1e1e1e"
    )

figure.update_xaxes(
    gridcolor="#1f292f",
    showgrid=True, fixedrange=True,rangeslider=dict(visible=False)
)
figure.update_yaxes(
    gridcolor="#1f292f",
    showgrid=True,
)


figure.show()


```
